```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('summer2016.csv')
```

```python
df.head(7)
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head()
```

```python
fig, ax = plt.subplots()
ax.hist(df["Age"])
ax.set_xlabel("Age")
ax.set_ylabel("# of observations")
plt.show()
```

```python
df_wrestling = df[df['Sport'] == 'Wrestling']
df_gymnastics = df[df['Sport'] == 'Gymnastics']
```

```python
fig, ax = plt.subplots()
ax.hist(df_wrestling["Height"], label='Wrestling')
ax.hist(df_gymnastics["Height"], label='Gymnastics')
ax.set_xlabel("Height")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(df_wrestling["Height"], label='Wrestling', histtype="step")
ax.hist(df_gymnastics["Height"], label='Gymnastics', histtype="step")
ax.set_xlabel("Height")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```
