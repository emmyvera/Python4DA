```python
import pandas as pd
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.show()
```

```python
df = pd.read_csv('austin_weather.csv')
```

```python
df['DATE']
```

```python
df["MLY-TAVG-NORMAL"]
```

### Line Graph

```python
fig, ax = plt.subplots()
ax.plot(df["DATE"], df["MLY-TAVG-NORMAL"])
plt.show()
```

```python
a = [1,2,3,4,5]
b = [1,2,3,4,5]

fig, ax = plt.subplots()
ax.plot(a, b)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df["DATE"], df["MLY-TAVG-NORMAL"])
ax.plot(df["DATE"], df["MLY-TMAX-NORMAL"])
plt.show()
```

```python
a = [1,2,3,4,5]
b = [1,2,3,4,5]
c = [2,4,6,8,10]

fig, ax = plt.subplots()
ax.plot(a, b)
ax.plot(a, c)
plt.show()
```

### Added Pointers

```python
fig, ax = plt.subplots()
ax.plot(df["DATE"], df["MLY-TAVG-NORMAL"], marker='o')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df["DATE"], df["MLY-TAVG-NORMAL"], marker = 'o')
ax.plot(df["DATE"], df["MLY-TMAX-NORMAL"], marker = 'v')
plt.show()
```

### Change Line Style

```python
fig, ax = plt.subplots()
ax.plot(df["DATE"], df["MLY-TAVG-NORMAL"], 
        marker='o', linestyle="--")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df["DATE"], df["MLY-TAVG-NORMAL"], 
        marker='o', linestyle='')
plt.show()
```

### Change Color

```python
fig, ax = plt.subplots()
ax.plot(df["DATE"], df["MLY-TAVG-NORMAL"], 
        marker='o', linestyle="--", color='orange')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df["DATE"], df["MLY-TAVG-NORMAL"], 
        marker = 'o', color='k')
ax.plot(df["DATE"], df["MLY-TMAX-NORMAL"], 
        marker = 'v', color='g')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(df["DATE"], df["MLY-TAVG-NORMAL"], 
        marker = 'o', color='k')
ax.plot(df["DATE"], df["MLY-TMAX-NORMAL"], 
        marker = 'v', color='g')

ax.set_xlabel("Time (months)")
ax.set_ylabel("Temperature")
plt.show()
```
